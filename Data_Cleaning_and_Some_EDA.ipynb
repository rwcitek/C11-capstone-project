# Data Cleaning

##Change column dtypes to appropriate dtypes

Columns to change:
- LAST EDITED DATE (obj -> datetime)
- OBSERVATION COUNT (obj - int)
- OBSERVATION DATE (obj -> datetime)
- TIME OBSERVATION STARTED (obj -> time)


```python
#change X's in OBSERVATION COUNT to 1's
df['OBSERVATION COUNT'].replace(to_replace=dict(X=1), inplace=True)
```

```python
df = df.astype({'OBSERVATION COUNT':'int'})
```

```python
df['LAST EDITED DATE'] = pd.to_datetime(df['LAST EDITED DATE'])
df['OBSERVATION DATE'] = pd.to_datetime(df['OBSERVATION DATE'])
df['TIME OBSERVATIONS STARTED'] = pd.to_datetime(df['TIME OBSERVATIONS STARTED'])
```
